```python
import ast
def compute_genre(obj):
    L=[] # List
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
mov['genres'] = mov['genres'].apply(compute_genre)
```

```python
#movies.drop('genre')
mov.head()
```

```python
mov['keywords']=mov['keywords'].apply(compute_genre)
```

```python
def compute_cast(obj):
    L=[] # List
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
           L.append(i['name'])
           counter+=1 
        else:
            break;
    return L  
```

```python
mov['cast']=mov['cast'].apply(compute_cast)
```

```python
mov.head()
```

```python
def compute_director(obj):
    L=[] # List
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
           L.append(i['name'])
           break;
    return L  
```

```python
mov['crew']=mov['crew'].apply(compute_director)
```

```python
mov['overview']=mov['overview'].apply(lambda x:x.split())
```

```python
mov['genres']=mov['genres'].apply(lambda x:[i.replace(" ","") for i in x])
mov['keywords']=mov['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
mov['cast']=mov['cast'].apply(lambda x:[i.replace(" ","") for i in x])
mov['crew']=mov['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
mov.head()
```

```python
mov['tag'] = mov['overview']+mov['genres']+mov['keywords']+mov['crew']
```
